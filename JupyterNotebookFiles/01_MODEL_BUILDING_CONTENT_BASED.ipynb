```python
from nltk.stem.porter import PorterStemmer
ps  = PorterStemmer() 
final_df['tags'] = final_df['tags'].apply(lambda x : [ps.stem(i) for i in x])
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x: " ".join(x).lower())
```

```python
"""
    Vectorization
    frequency of most frequent 6000 words for each movie
"""

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=800, stop_words='english')
vectors = cv.fit_transform(final_df['tags']).toarray()
```

# COSINE SIMILLARITIES

```python
"""
    Cosine Similarites
"""

from sklearn.metrics.pairwise import cosine_similarity
simillarity = cosine_similarity(vectors)
```

```python
import difflib
```

```python
movie_list = final_df['title'].values.tolist()
```

# FUNCTION TO RECOMEND MOVIES

```python
def recommend(movie_name):
    movie_name = difflib.get_close_matches(movie_name,movie_list)[0]
    
    movie_index = df[df['title'] == movie_name].index[0]
    cosine_distance = simillarity[movie_index]

    required_movies = sorted(enumerate(cosine_distance),reverse=True, key=lambda x:x[1])[0:20]
    
    data = []
    
    for i in required_movies:
        item = []
        item.append(df[df['imdb_id'] == df.iloc[i[0]]['imdb_id']]['title'].values[0])
        item.append(" ".join(df[df['imdb_id'] == df.iloc[i[0]]['imdb_id']]['summary'].values[0]))
        item.append(df[df['imdb_id'] == df.iloc[i[0]]['imdb_id']]['actors'].values[0])
        item.append(df[df['imdb_id'] == df.iloc[i[0]]['imdb_id']]['wiki_link'].values[0])
        item.append(df[df['imdb_id'] == df.iloc[i[0]]['imdb_id']]['poster_path'].values[0])
        data.append(item)
        
    print(data)
```

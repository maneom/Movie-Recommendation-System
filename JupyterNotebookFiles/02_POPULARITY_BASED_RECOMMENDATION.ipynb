# MAIN APPROACH FOR RECOMMENDING POPULAR BASED MOVIES

```python
"""
    CONSIDERING POPULAR MOVIES AS BET RATE 6 TO 8
    AND VOTES 30000 TO 50000
"""
pass
```

```python
df.drop_duplicates(inplace=True)
```

```python
popular_movies = df[(df['imdb_rating'] > 6) & ((df['imdb_votes']>= 30000))].sort_values(by=['imdb_votes'], ascending=False)
```

## COUNT VECTORIZER

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=23)
genres = cv.fit_transform(df['genres']).toarray()
```

```python
genres.shape
```

```python
diff_genres_name = cv.get_feature_names_out()
```

```python
genres
```

```python
df.iloc[0]['genres']
```

```python
genres[0]
```

```python
genres_data = pd.DataFrame(data=genres, index=df['title'].values , columns=diff_genres_name)
```

```python
genres_data.drop(['drama','fi','sci'],axis=True, inplace=True)
```

```python
genres_data.drop(['western'],axis=True, inplace=True)
```

```python
genres_data.drop(['musical'],axis=True, inplace=True)
```

## FUNCTION FOR PREDICTION BASED ON GENRES

```python
def based_on_genre(value):
    mov_list = genres_data[genres_data[value]==1].index
    
    for i in mov_list:
        print(df[df['title'] == mov_list[0]]['poster_path'].values)
        print(df[df['title'] == mov_list[0]]['title'].values)
        print(df[df['title'] == mov_list[0]]['summary'].values)
        print(list(df[df['title'] == mov_list[0]]['actors'].values))
        print(df[df['title'] == mov_list[0]]['wiki_link'].values)
        print("\n\n")
```
